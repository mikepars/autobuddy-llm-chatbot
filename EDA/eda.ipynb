# **Exploratory Data Analysis** 

Tujuan cardinality mengetahui nilai - nilai dalam dataset untuk nantinya kolom - kolom itu akan kita lakukan pembersihan yang tujuannya untuk mempermudah saat melakukan modelling. Berikut adalah tahap - tahapannya:
* Melihat outlier
* Melihat data duplikat
* Melihat data yang memilki nilai "NULL"
* Melihat nama - nama dalam kolom

```python
# Membuat list untuk menyimpan data
cardinality_dict = []

# Menghitung dan menyimpan cardinality untuk setiap kolom
for column in df_eda.columns:
    cardinality = df_eda[column].nunique()
    cardinality_dict.append({'Column Name': column, 'Cardinality': cardinality})

# Membuat DataFrame dari list
cardinality_df = pd.DataFrame(cardinality_dict)

# Menyortir DataFrame berdasarkan kardinalitas (Cardinality) secara menurun
sorted_cardinality_df = cardinality_df.sort_values(by='Cardinality', ascending=False)

# Menampilkan DataFrame yang telah disortir
print("Cardinality Kolom")
print(sorted_cardinality_df.to_string())
```

- Terlihat fitur yang memiliki kardinalitas tinggi yaitu yang memiliki nilai unik tinggi. Fitur yang memiliki kardinalitas tinggi yaitu **mileage**, kemudian diikuti oleh **instalment**, dan **price** karena memiliki nilai unique ratusan ribu sampai ribuan.

**Melihat kolerasi antar kolom kategorikal**

```python
# Filter only categorical columns (object dtype)
categorical_df = df_eda.select_dtypes(include=['object'])

# Function to calculate Cramér's V
def cramers_v(x, y):
    contingency_table = pd.crosstab(x, y)
    chi2, p, dof, expected = chi2_contingency(contingency_table)
    n = contingency_table.sum().sum()
    return np.sqrt(chi2 / (n * (min(contingency_table.shape) - 1)))

# Create a matrix to store Cramér's V values
cramers_v_matrix = pd.DataFrame(index=categorical_df.columns, columns=categorical_df.columns)

# Calculate Cramér's V for each pair of categorical variables
for col1 in categorical_df.columns:
    for col2 in categorical_df.columns:
        if col1 != col2:
            cramers_v_matrix.loc[col1, col2] = cramers_v(categorical_df[col1], categorical_df[col2])
        else:
            cramers_v_matrix.loc[col1, col2] = 1  # Cramér's V with itself is 1

# Convert to float for heatmap
cramers_v_matrix = cramers_v_matrix.astype(float)

# Generate heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(cramers_v_matrix, annot=True, cmap='coolwarm', vmin=0, vmax=1)
plt.title('kolerasi antar kolom kategorikal')
plt.show()
```

Gambar ini menunjukkan peta korelasi antara berbagai kolom kategorikal dalam dataset mobil. Korelasi diukur menggunakan koefisien korelasi Cramér's V, yang ditampilkan dalam bentuk heatmap. Nilai korelasi berkisar antara 0 dan 1, di mana:

- Nilai 1 menunjukkan korelasi sempurna.
- Nilai 0 menunjukkan tidak ada korelasi.

1. **Korelasi yang Sangat Tinggi:**
   - **Nama Mobil dan Brand:** Korelasi sempurna sebesar 1, menunjukkan bahwa setiap nama mobil berkorelasi secara sempurna dengan merek mobil. Ini masuk akal karena setiap model mobil biasanya hanya terkait dengan satu merek.
   - **Nama Mobil dan Transmisi:** Korelasi sebesar 0.72, menunjukkan bahwa tipe transmisi tertentu sering dikaitkan dengan model mobil tertentu.

2. **Korelasi berimbang:**
   - **Nama Mobil dan Plate Type:** Korelasi sebesar 0.57, menunjukkan bahwa tipe plat nomor mobil memiliki hubungan beriimbang dengan model mobil.
   - **Nama Mobil dan Location:** Korelasi sebesar 0.56, menunjukkan bahwa lokasi di mana mobil dijual atau terdaftar memiliki hubungan imbang dengan model mobil.
   - **Brand dan Transmisi:** Korelasi sebesar 0.42, menunjukkan bahwa tipe transmisi tertentu sering dikaitkan dengan merek mobil tertentu.

3. **Korelasi Rendah atau Tidak Signifikan:**
   - **Brand dan Location:** Korelasi sebesar 0.14, menunjukkan bahwa merek mobil tidak banyak berhubungan dengan lokasi.
   - **Location dan Transmisi:** Korelasi sebesar 0.16, menunjukkan bahwa lokasi tidak banyak berhubungan dengan tipe transmisi mobil.
   - **Transmisi dan Plate Type:** Korelasi sangat rendah sebesar 0.018, menunjukkan bahwa tipe transmisi hampir tidak ada hubungannya dengan tipe plat nomor mobil.
   - **Location dan Plate Type:** Korelasi sebesar 0.10, menunjukkan bahwa lokasi tidak banyak berhubungan dengan tipe plat nomor mobil.
   - **Brand dan Plate Type:** Korelasi sebesar 0.17, menunjukkan bahwa merek mobil memiliki sedikit hubungan dengan tipe plat nomor mobil.

Dari analisis ini, kita dapat menyimpulkan bahwa nama mobil sangat berkorelasi dengan merek dan tipe transmisinya. Selain itu, ada korelasi moderat antara nama mobil dengan tipe plat nomor dan lokasi. Korelasi antara merek dan tipe transmisi juga cukup signifikan. Namun, hubungan antara lokasi dengan merek, tipe transmisi, dan tipe plat nomor mobil cenderung rendah atau tidak signifikan.
