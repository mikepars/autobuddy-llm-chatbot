# 1. Import Libraries

```python
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
import time

import numpy as np
import pandas as pd
```

# 2. Web Scrapping

```python
%%time
# Setup driver (pastikan chromedriver sudah terinstall dan di PATH)
driver = webdriver.Chrome()
```

```python
# FAQ sub-category
subcat = ['jual', 'beli', 'dealer', 'Tukar%20Tambah%20Mobil']

question = []
answer = []

for cat in subcat:
    url = f"https://www.carsome.id/faqs#{cat}"
    driver.get(url)
    
    time.sleep(5)
    
    # driver.execute_script("window.scrollBy(0, 300)")
    
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    
    faqs = soup.findAll("div", {"class": "v-expansion-panel"})
    
    for faq in faqs:
        
        answ = []
    
        title = faq.find("h3", class_ = None)
        ans = faq.find("div", {"class": "faq__detail"})
        
        for content in ans.findAll():
            answ.append(content.text)
            
        if title is not None:
            question.append(title.text)
            print(f"Appended {title.text}")
        else:
            question.append(None)
        
        if ans is not None:
            answer.append(answ)
            print(f"Appended {answ}")
        else:
            answer.append(None)
        
df = pd.DataFrame({'Question': question, 'Answer': answer})


driver.quit()
```
